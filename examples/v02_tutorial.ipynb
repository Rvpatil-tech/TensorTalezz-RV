# TensorTalezz-RV v0.2.0 Tutorial

Welcome to the official tutorial! Here we will demonstrate the new features of v0.2.0:
1. **Responsible AI**: Detecting bias and fairness.
2. **Deployment**: Exporting and serving your model.
3. **Reinforcement Learning**: Running a quick GridWorld.

```python
import pandas as pd
import numpy as np
import tensor_talezz_rv
from tensor_talezz_rv import detect_bias, fit, export_model, serve_model
from tensor_talezz_rv import rl_demo, reward_curve
```

## 1. Responsible AI: Detecting Bias
Let's create a synthetic dataset where 'gender' unfairly influences the 'hired' variable.

```python
np.random.seed(42)
n = 100
df = pd.DataFrame({
    'age': np.random.randint(20, 50, n),
    'gender': np.random.choice(['Male', 'Female'], n),
})
# Create bias: Males are more likely to be hired
df['hired'] = np.where(df['gender'] == 'Male', 
                       np.random.choice([0, 1], p=[0.2, 0.8], size=n), 
                       np.random.choice([0, 1], p=[0.7, 0.3], size=n))

# Detect bias before training
report = detect_bias(df, target='hired', sensitive_feature='gender')
```

## 2. Deployment: Exporting and Serving
Now we train on this data, export it, and generate a web server script automatically.

```python
model, _ = fit(df, target='hired', verbose=False)

# Export to disk
model_path = export_model(model, filepath='hiring_model.joblib')

# Generate an API deployment script
api_script = serve_model(model_path=model_path, port=8000)
```

## 3. Reinforcement Learning Basics
Let's take a break and watch a Q-Learning agent solve a maze.

```python
agent = rl_demo(episodes=100, display=True)
reward_curve(agent)
```
